```python
import numpy as np
import pandas as pd
```

```python
# FINDING MISSING DATA
```

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [np.nan, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

```python
# REMOVING MISSING DATA
```

```python
df.dropna()
```

```python
df
```

```python
df.dropna(thresh=3)
```

```python
df
```

```python
# FILLING THE MISSING DATA
```

```python
df
```

```python
df.fillna(0)
```

```python
df
```

```python
values = {"A": 0, "B": 10, "C": 100, "D": 1000}
df.fillna(value=values)
```
