**THE DISTRIBUTION PLOTS ARE USED TO VISUALIZE QUANTITATIVE DATA(SALARY, AGE)**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
df["size"].unique()
```

```python
# DISTPLOT / HISTPLOT
```

```python
sns.displot(df["total_bill"])
```

```python
sns.displot(df["total_bill"], bins=10)
```

```python
sns.displot(df["total_bill"], bins=20, kde=True) # Kernel Density Estimator (KDE)
```

```python
sns.histplot(df["total_bill"])
```

```python
plt.subplot(1,2,1)
sns.histplot(df["total_bill"], bins=20, kde=True)
plt.subplot(1,2,2)
sns.histplot(df["tip"], bins=20, kde=True)
```

```python
# JOINTPLOT
```

```python
sns.jointplot(x="total_bill", y="tip", data=df, kind="scatter")
```

```python
sns.jointplot(x="total_bill", y="tip", data=df, kind="kde")
```

```python
sns.jointplot(x="total_bill", y="tip", data=df, kind="hist")
```

```python
sns.jointplot(x="total_bill", y="tip", data=df, kind="hex")
```

```python
sns.jointplot(x="total_bill", y="tip", data=df, kind="reg")
```

```python
sns.jointplot(x="total_bill", y="tip", data=df, kind="resid")
```

```python
# PAIRPLOT
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue="sex")
```

```python
sns.pairplot(df, hue="size", palette="rainbow")
```

```python
# RUGPLOT
```

```python
sns.rugplot(df)
```

```python
sns.rugplot(df["total_bill"])
```
